```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('3.01.+Country+clusters.csv')
data
```

```python
x = data.iloc[:,1:3]
x
```

```python
# Apply K-Means clustering using below methods
# KMeans(2) Number in object shows how much clusters you wants 
kmeans = KMeans(3)

#Fit x/inputs to kmeans
kmeans.fit(x)
```

```python
#Predict the clusters using fit_predict function
clusters = kmeans.fit_predict(x)
clusters
```

```python
data['clusters'] = clusters
data
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data['clusters'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
